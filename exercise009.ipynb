# Exploring a larger array

```python
import matplotlib.pyplot as plt
import numpy as np
```

Load the data

```python
data2 = np.load('datasets/009ExerciseFile2.npy')
```

```python
data2.shape
```

Plot the 5th row

```python
plt.plot(data2[4,:])
```

Plotting the first 20 rows of all columns.

Added the semicolon to ignore the list of objects drawn.

```python
plt.plot(data2[0:20,:]);
```

Sum all the rows within a column

```python
columnsum = data2.sum(0)
```

```python
columnsum.shape
```

```python
plt.plot(columnsum)
```

```python
columnmean = data2.mean(0)
```

```python
plt.plot(columnmean)
```

There's no pattern in the columns, so now I'm going to sum all the columns within a row.

```python
rowsum = data2.sum(1)
```

```python
rowsum.shape
```

```python
plt.plot(rowsum)
```

```python
rowmean = data2.mean(1)
```

```python
rowmean.shape
```

### Found the Pattern!

```python
plt.plot(rowmean)
plt.title("Averages of the Data Rows")
```

```python
import seaborn as sns
```

Smoothed out the plot using kde

```python
sns.displot(rowmean, kind='kde')
plt.title("Average of the Data Rows")
```

It seems that some of the rows have much higher means than the others!
