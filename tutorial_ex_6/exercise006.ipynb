# Exercise 6 Reading and Writing Data Files

Import the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the data "006ExerciseFile.csv" into a pandas data frame

```python
myData = pd.read_csv("datasets/006ExerciseFile.csv")
```

Make histograms of the 2 distributions

```python
myData.hist()
```

Prettier histograms

```python
sns.histplot(myData, multiple="layer", element="step")
```

Make pretty KDE plots of the two distributions

```python
sns.kdeplot(myData["frzr4"], color="#F13FD7", fill=True, alpha=0.2, label="Freezer 4")
sns.kdeplot(myData["frzr2"], color="#3F67F1", fill=True, alpha=0.2, label="Freezer 2")
plt.legend()
```

Summarize the data

```python
mySummary = myData.describe()
```

```python
print(mySummary)
```

Creating a new file with the data summary

```python
mySummary.to_csv("ex6Summary.csv")
```

Read the summary back in and confirm the read was successful

```python
mySummary2 = pd.read_csv("ex6Summary.csv")
```

```python
display(mySummary2)
```
