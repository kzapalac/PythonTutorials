# Project 2: Data Visualization
For this project, I compared wages across genders to see if there was a wage gap between males and females. Additionally, I looked to see if age or education levels impacted the wage gap between males and females.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
md = pd.read_csv("datasets/wageGenderEduAge.csv")
```

```python
display(md)
```

## Comparing wages per gender 

**Use tutorial and exercise 7 to create a distribution plot**

```python
sns.set_style('whitegrid')
sns.plotting_context() # tells me what the pre-set stuff is, and I can change it with sns.set_context
```

```python
sns.set_context('notebook', font_scale=1.5) #notebook is the default
```

Histogram of the wage distribution for males vs. females

```python
sns.displot(md, x='wage', hue='gender', kind='hist', 
            alpha=0.3, height=5, aspect=2, palette='Set1')
```

KDE of the wage distribution for females vs. males

```python
sns.set_context(font_scale=2)
sns.displot(md, x='wage', hue='gender', kind='kde', 
            fill=True, alpha=0.3, palette="Set1", height=5, aspect=2)
```

```python
sns.displot(md, x="wage", hue="gender", kind="hist", kde=True, 
            palette='Set1', height=5, aspect=2)
plt.xlabel("Wage")
plt.title("Wage Distributions by Gender")
```

Boxplot of the wage distribution for females vs. males

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=md, x='wage', y='gender', palette="Set1")
sns.stripplot(data=md, x='wage', y='gender', alpha=0.4, color='black')
plt.title("Female vs. Male Wage Distribution")
plt.xlabel("Wage")
plt.ylabel("Gender")
```

```python
md.groupby('gender').describe()
```

## Effect of education on pay gap

**Use tutorial and exercise 8 to create a plot showing a relationship**

Scatterplot of the relationship

```python
sns.lmplot(data=md, x='education', y='wage', hue='gender', 
           height=6, aspect=1.5, scatter_kws = {"s" : 20}, x_jitter=0.1)
plt.title("Relationship Between Education and Wage per Gender")
plt.xlabel("Education")
plt.ylabel("Wage")
```

```python
sns.jointplot(data=md, x="education", y="wage", kind="kde", 
              fill=True, hue='gender', alpha=0.6, palette='Set1')
plt.xlabel('Education')
plt.ylabel('Wage')
plt.title('Joint Plot of Wages by Age and Gender')
```

## Effect of age on pay gap

```python
sns.lmplot(data=md, x='age', y='wage', hue='gender',
           palette='Set1', height=6, aspect=1.5, scatter_kws = {"s" : 20})
plt.title("Relationship Between Age and Wage per Gender")
plt.xlabel("Age")
plt.ylabel("Wage")
```

```python
sns.jointplot(data=md, x="age", y="wage", kind="kde", 
              fill=True, hue='gender', alpha=0.6, palette='Set1')
plt.xlabel('Age')
plt.ylabel('Wage')
plt.title('Joint Plot of Wages by Age and Gender')
```
