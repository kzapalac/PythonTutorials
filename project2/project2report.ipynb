# Project 2: Data Visualization
For this project, I compared wages across genders to see if there was a wage gap between males and females. Additionally, I looked to see if age or education levels impacted the wage gap between males and females.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
md = pd.read_csv("datasets/wageGenderEduAge.csv")
```

```python
display(md)
```

## Comparing wages per gender 

**Use tutorial and exercise 7 to create a distribution plot**

Histogram of the wage distribution for males vs. females

```python
sns.displot(md, x='wage', hue='gender', kind='hist', alpha=0.3)
```

KDE of the wage distribution for females vs. males

```python
sns.displot(md, x='wage', hue='gender', kind='kde', fill=True, alpha=0.3)
```

Boxplot of the wage distribution for females vs. males

```python
sns.boxplot(data=md, x='wage', y='gender')
sns.stripplot(data=md, x='wage', y='gender', alpha=0.4)
```

## Effect of education on pay gap

**Use tutorial and exercise 8 to create a plot showing a relationship**

Scatterplot of the relationship

```python
sns.lmplot(data=md, x='education', y='wage', hue='gender', scatter_kws = {"s" : 5})
```

## Effect of age on pay gap

```python
sns.lmplot(data=md, x='age', y='wage', hue='gender', scatter_kws = {"s" : 5})
```
